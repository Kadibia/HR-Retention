```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("HR_comma_sep.csv")
df
```

```python
df['salary'].describe()
```

```python
salary_map = {
    'low': 1, 
    'medium': 2, 
    'high':3
}
df['Salary'] = df['salary'].map(salary_map)

department_map ={
    'sales': 1,
    'accounting': 2,
    'hr': 3,
    'technical': 4,
    'support': 5,
    'management': 6,
    'IT': 7,
    'product_mng': 8,
    'RandD': 9,
    'marketing': 10
}
df['department'] = df['Department'].map(department_map)

df
```

```python
plt.figure(figsize=(8,6))
left_counts = df['left'].value_counts()
plt.bar(left_counts.index, left_counts.values, tick_label=['Stayed', 'left'])
plt.title('Distribution of Employee Retention (Left vs Stayed)')
plt.xlabel('Left')
plt.ylabel('Count')
plt.show();
```

```python
plt.figure(figsize=(8, 6))
satisfaction_level_means = df.groupby('left')['satisfaction_level'].mean()
plt.bar(satisfaction_level_means.index, satisfaction_level_means.values, tick_label=["Stayed", "left"])
plt.title("Satisfaction Level vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Average Satisfaction Level")
plt.show()
```

```python
plt.figure(figsize=(8,6))
last_evaluation_means = df.groupby('left')['last_evaluation'].mean()
plt.bar(last_evaluation_means.index, satisfaction_level_means.values, tick_label=["Stayed", "left"])
plt.title("Last Evaluation vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Average Last Evaluation Level")
plt.show();
```

```python
plt.figure(figsize=(8,6))
number_project_means = df.groupby('left')['number_project'].mean()
plt.bar(number_project_means.index, number_project_means.values, tick_label=["Stayed", "Left"])
plt.title("Number of Project vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Average Number of Project")
plt.show();
```

```python
plt.figure(figsize=(8,6))
average_monthly_hours_means = df.groupby('left')['average_montly_hours'].mean()
plt.bar(average_monthly_hours_means.index, average_monthly_hours_means.values, tick_label=["Stayed", "Left"])
plt.title("Average Monthly Hours vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Average Monthly Hours")
plt.show();
```

```python
plt.figure(figsize=(8,6))
time_spend_company_means = df.groupby("left")["time_spend_company"].mean()
plt.bar(time_spend_company_means.index, time_spend_company_means.values, tick_label=["Stayed", "Left"])
plt.title("Timespent in Company vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Time Spent in Company")
plt.show();
```

```python
plt.figure(figsize=(8,6))
Work_accident_means = df.groupby("left")["Work_accident"].mean()
plt.bar(Work_accident_means.index, Work_accident_means.values, tick_label=["Stayed", "Left"])
plt.title("Work accident vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Work accident in Company")
plt.show();
```

```python
plt.figure(figsize=(8,6))
promotion_last_5years_means = df.groupby("left")["promotion_last_5years"].mean()
plt.bar(promotion_last_5years_means.index, promotion_last_5years_means.values, tick_label=["Stayed", "Left"])
plt.title("Promotion in The Last 5 Years vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Promotion in The Last 5 Years")
plt.show();
```

```python
plt.figure(figsize=(8,6))
Department_means = df.groupby("left")["department"].mean()
plt.bar(Department_means.index, Department_means.values, tick_label=["Stayed", "Left"])
plt.title("Department vs Employee Retention")
plt.xlabel("Left")
plt.ylabel("Department")
plt.show();
```

```python
plt.figure(figsize=(12, 8))
departments = df.groupby(['Department', 'left']).size().unstack().fillna(0)
departments.plot(kind='bar', stacked=True)
plt.title('Department vs Employee Retention')
plt.xlabel('Department')
plt.ylabel('Count')
plt.show();
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Salary']],df.left, train_size=0.9)
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
